# Linear Object Skeletonization

This is an implementation of the linear object skeletonization algorithm from Taohan Wang and Yuji Yamakawa
https://www.mdpi.com/1424-8220/23/12/5721

```python
import matplotlib.pyplot as plt
import numpy as np

import skimage.io as io
from skimage.filters import threshold_otsu, sobel
from skimage.color import rgb2gray
from skimage.segmentation import watershed

import cv2
from scipy import ndimage
```

```python
i= 4
img_original =  io.imread(f'./examples/{i}_knot.png')[...,:3]      # Gray image, rgb images need pre-conversion
img_gray = rgb2gray(img_original)
```

## Binarization and Segmentation

```python
elevation_map = sobel(img_gray)
plt.imshow(elevation_map)
```

```python
markers = np.zeros_like(img_gray)
markers[img_gray<30/255]=1
markers[img_gray>150/255]=2
markers = markers.astype(np.int32)
plt.imshow(markers)
```

```python
segment = watershed(elevation_map, markers)-1
plt.imshow(segment)
```
